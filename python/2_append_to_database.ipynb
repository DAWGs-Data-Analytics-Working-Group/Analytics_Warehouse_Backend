```python
import pandas as pd
import numpy as np
import os
# import gc
from typing import Iterator, Union, List, Optional
from pathlib import Path
import shutil
```

```python
# Read the CSV and extract column names as a list
cols = pd.read_csv("../columns.csv")["Column Names"].tolist()
parquet_files = [f for f in os.listdir('../planning_data/ped_report_parquet') if f.endswith('.parquet')]
parquet_files
```

```python
dfs = []
for file in parquet_files:
    file_path = os.path.join('../planning_data/ped_report_parquet', file)
    arch_path = os.path.join('../planning_data/ped_report_parquet/archive', file)
    try:
        # Read the Excel file with specified columns
        df = pd.read_parquet(file_path, engine="pyarrow")
        print(f"Successfully read: {file}")
        dfs.append(df)
        print(f"Successfully appended: {file}")
        # Source path 
        dest = shutil.move(file_path, arch_path) 
    except Exception as e:
        print(f"Error reading {file}: {e}")
else:
    print("No valid data to save.")

if dfs:
    # Concatenate all DataFrames into one
    combined_df = pd.concat(dfs, ignore_index=True)
    
else:
    print("No Parquet files to process.")
```

```python
# combined_df
df = combined_df.drop_duplicates().reset_index(drop = True)
df
```

```python
df_database = pd.read_parquet("../planning_data/database/database_cleaned.parquet", engine="pyarrow")
df_database
```

```python
df["Time_Bucket"] = df["From"].dt.floor("30T")  # "30T" means 30-minute interval
grouped_df = df.groupby("Time_Bucket").sum(numeric_only=True).reset_index()
long_df = grouped_df.melt(id_vars=["Time_Bucket"], var_name="Location", value_name="Count")
long_df
```

```python
sensor_mapping = pd.read_csv("../../Planning/Sensor_Mapping.csv")
sensor_mapping
```

```python
long_df_join = long_df.merge(sensor_mapping, how = "left", left_on = "Location", right_on = "Sensor Names")
long_df_join_clean = long_df_join[["Time_Bucket","Sensor Names","Count","Hub In/Out","Location - Monthly Ped Report","Zone Number"]].dropna()
long_df_join_clean = long_df_join_clean.groupby(["Time_Bucket","Hub In/Out","Location - Monthly Ped Report","Zone Number"])["Count"].sum().reset_index()
re_concatenate = [df_database,long_df_join_clean]
combined_df = pd.concat(re_concatenate, ignore_index=True)
combined_df = combined_df.drop_duplicates()
parquet_filename = f"../planning_data/database/database_cleaned.parquet"
combined_df.to_parquet(parquet_filename, engine="pyarrow", index=False)
```
