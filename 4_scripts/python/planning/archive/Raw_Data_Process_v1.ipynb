```python
import pandas as pd
import gc
from pathlib import Path
from typing import Union, List, Iterator, Optional

def read_excel_files_efficiently(
    directory: Union[str, Path],
    pattern: str = "*.xlsx",
    sheet_name: str = "Data",
    nrows: Optional[int] = None,
    usecols: Optional[List[str]] = None,
    chunk_size: int = 10000
) -> Iterator[pd.DataFrame]:
    """
    Generator function to read multiple Excel files efficiently in chunks.
    
    Args:
        directory: Path to directory containing Excel files
        pattern: File pattern to match (e.g., "*.xlsx", "data_*.xls")
        sheet_name: Name of the sheet to read
        nrows: Number of rows to read (optional)
        usecols: List of columns to read. If None, reads all columns
        chunk_size: Number of rows to process at a time
    
    Yields:
        DataFrame chunks from each matching Excel file
    """
    directory = Path(directory)
    
    for excel_file in directory.glob(pattern):
        try:
            # Read the Excel file with memory optimization options
            df = pd.read_excel(
                excel_file,
                usecols=usecols,
                sheet_name=sheet_name,
                nrows=nrows,
                engine='openpyxl'
            )
            
            # Process in smaller chunks
            num_chunks = len(df) // chunk_size + 1
            
            for i in range(num_chunks):
                start_idx = i * chunk_size
                end_idx = min((i + 1) * chunk_size, len(df))
                
                chunk = df.iloc[start_idx:end_idx].copy()
                yield chunk, excel_file.name  # Also yield filename for tracking
                
                # Clear chunk from memory
                del chunk
                gc.collect()
            
            # Clear full dataframe from memory
            del df
            gc.collect()
            
        except Exception as e:
            print(f"Error reading {excel_file}: {str(e)}")
            continue

def process_excel_files(
    directory: Union[str, Path],
    output_file: Union[str, Path],
    columns_to_read: Optional[List[str]] = None,
    pattern: str = "*.xlsx",
    sheet_name: str = "Data",
    chunk_size: int = 10000,
    process_func: Optional[callable] = None
) -> None:
    """
    Process multiple Excel files and save results efficiently.
    
    Args:
        directory: Directory containing Excel files
        output_file: Path to save processed results
        columns_to_read: List of columns to read from Excel files
        pattern: File pattern to match
        sheet_name: Name of the sheet to read
        chunk_size: Size of chunks to process at a time
        process_func: Optional function to apply to each chunk before saving
    """
    output_file = Path(output_file)
    
    # Create output directory if it doesn't exist
    output_file.parent.mkdir(parents=True, exist_ok=True)
    
    # Initialize tracking variables
    files_processed = 0
    total_rows = 0
    
    # Process files in chunks
    for chunk, filename in read_excel_files_efficiently(
        directory,
        pattern=pattern,
        sheet_name=sheet_name,
        usecols=columns_to_read,
        chunk_size=chunk_size
    ):
        try:
            # Apply custom processing if provided
            if process_func:
                chunk = process_func(chunk)
            
            # Append to results, using mode='a' for append after first write
            mode = 'w' if files_processed == 0 and total_rows == 0 else 'a'
            header = files_processed == 0 and total_rows == 0
            
            chunk.to_csv(output_file, mode=mode, header=header, index=False)
            
            # Update tracking
            total_rows += len(chunk)
            if mode == 'w':
                files_processed += 1
            
            # Progress update
            print(f"Processed {len(chunk)} rows from {filename}")
            
            # Clear processed chunk from memory
            del chunk
            gc.collect()
            
        except Exception as e:
            print(f"Error processing chunk from {filename}: {str(e)}")
            continue
    
    print(f"\nProcessing complete:")
    print(f"Files processed: {files_processed}")
    print(f"Total rows processed: {total_rows}")
    print(f"Results saved to: {output_file}")

# Example usage
if __name__ == "__main__":
    # Example configuration
    input_dir = Path("../Planning/Monthly Ped Reports")
    output_file = Path("processed_results.csv")
    
    # Optional: Define custom processing function
    def custom_processing(df: pd.DataFrame) -> pd.DataFrame:
        # Add your custom processing logic here
        # For example:
        df['processed_date'] = pd.Timestamp.now()
        return df
    
    # Process files
    process_excel_files(
        directory=input_dir,
        output_file=output_file,
        columns_to_read=None,  # Specify columns if needed
        process_func=custom_processing

    )
```

```python
def read_excel_files_efficiently(
    directory: Union[str, Path],
    pattern: str = "*.xlsx",
    sheet_name: str = "Data",
    nrows: Optional[int] = None,
    usecols: Optional[List[str]] = None,
    chunk_size: int = 10000
) -> Iterator[pd.DataFrame]:
    """
    Generator function to read multiple Excel files efficiently in chunks.
    
    Args:
        directory: Path to directory containing Excel files
        pattern: File pattern to match (e.g., "*.xlsx", "data_*.xls")
        sheet_name: Name of the sheet to read
        nrows: Number of rows to read (optional)
        usecols: List of columns to read. If None, reads all columns
        chunk_size: Number of rows to process at a time
    
    Yields:
        DataFrame chunks from each matching Excel file
    """
    directory = Path(directory)
    
    for excel_file in directory.glob(pattern):
        try:
            # Read the Excel file with memory optimization options
            df = pd.read_excel(
                excel_file,
                usecols=cols,
                sheet_name=sheet_name,
                nrows=nrows,
                engine='openpyxl'
            )
            
            # Process in smaller chunks
            num_chunks = len(df) // chunk_size + 1
            
            for i in range(num_chunks):
                start_idx = i * chunk_size
                end_idx = min((i + 1) * chunk_size, len(df))
                
                chunk = df.iloc[start_idx:end_idx].copy()
                yield chunk, excel_file.name  # Also yield filename for tracking
                
                # Clear chunk from memory
                del chunk
                gc.collect()
            
            # Clear full dataframe from memory
            del df
            gc.collect()
            
        except Exception as e:
            print(f"Error reading {excel_file}: {str(e)}")
            continue

        
def process_excel_files(
    directory: Union[str, Path],
    output_file: Union[str, Path],
    columns_to_read: Optional[List[str]] = None,
    pattern: str = "*.xlsx",
    sheet_name: str = "Data",
    chunk_size: int = 10000,
    process_func: Optional[callable] = None
) -> None:
    """
    Process multiple Excel files and save results efficiently in Parquet format.
    
    Args:
        directory: Directory containing Excel files
        output_file: Path to save processed results
        columns_to_read: List of columns to read from Excel files
        pattern: File pattern to match
        sheet_name: Name of the sheet to read
        chunk_size: Size of chunks to process at a time
        process_func: Optional function to apply to each chunk before saving
    """
    output_file = Path(output_file)
    
    # Create output directory if it doesn't exist
    output_file.parent.mkdir(parents=True, exist_ok=True)
    
    # Initialize tracking variables
    files_processed = 0
    total_rows = 0
    
    # Process files in chunks
    for chunk, filename in read_excel_files_efficiently(
        directory,
        pattern=pattern,
        sheet_name=sheet_name,
        usecols=cols,
        chunk_size=chunk_size
    ):
        try:
            # Apply custom processing if provided
            if process_func:
                chunk = process_func(chunk)

            # Append to Parquet file
            # chunk.to_parquet(output_file, engine="pyarrow", compression="snappy", index=False, append=True)
# Inside the try block where Parquet saving happens
            try:
                # Choose the Parquet engine
                engine = "fastparquet"  # or "pyarrow"

                # Check if the file already exists
                try:
                    existing_df = pd.read_parquet(output_file, engine=engine)
                    df = pd.concat([existing_df, chunk], ignore_index=True)  # Concatenate the existing data with the new chunk
                except (FileNotFoundError, OSError):
                    df = chunk  # If file does not exist, initialize with the current chunk

                # Save to Parquet without 'append'
                df.to_parquet(output_file, engine=engine, index=False)

                print(f"Saved to {output_file}")

            except Exception as e:
                print(f"Error saving to Parquet: {e}")
                # Update tracking
                total_rows += len(chunk)
                files_processed += 1
                
                # Progress update
                print(f"Processed {len(chunk)} rows from {filename}")
                
                # Clear processed chunk from memory
                del chunk
                gc.collect()

            
        except Exception as e:
            print(f"Error processing chunk from {filename}: {str(e)}")
            continue
    
    print(f"\nProcessing complete:")
    print(f"Files processed: {files_processed}")
    print(f"Total rows processed: {total_rows}")
    print(f"Results saved to: {output_file}")
```

```python
if __name__ == "__main__":
    # Example configuration
    input_dir = Path("../Planning/Monthly Ped Reports")
    output_file = Path("processed_results.parquet")  # Change extension to .parquet
    
    # Optional: Define custom processing function
    def custom_processing(df: pd.DataFrame) -> pd.DataFrame:
        """Cleans and processes data before saving."""
        # Remove empty rows
        df = df.dropna(how="all")

        # Remove footer rows with unwanted text
        if not df.empty:
            df = df[~df.iloc[:, 0].astype(str).str.contains(r"^#|Weekly Average|Key Location|Hub In/Out", na=False)]

        # Ensure all date columns are converted properly
        if 'Date' in df.columns:
            df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # Convert invalid values to NaT

        # Convert numeric columns safely
        numeric_cols = ['To Time']  # Add more if needed
        for col in numeric_cols:
            if col in df.columns:
                df[col] = pd.to_numeric(df[col], errors='coerce')  # Convert invalid values to NaN
        
        # Add processing timestamp
        df['processed_date'] = pd.Timestamp.now()
        
        return df
    
    # Process files
    process_excel_files(
        directory=input_dir,
        output_file=output_file,
        columns_to_read=cols,  # Specify columns if needed
        process_func=custom_processing
    )
```
