```python
import os
import csv
import base64
from io import BytesIO
from PIL import Image

# === Configure these paths ===
photos_folder = dest_folder    # folder that contains your previously saved PNGs
output_folder = base_folder      # folder where you want the CSV to go
output_csv_file = "photos_base64.csv"   # CSV filename inside output_folder

# Target character limit for the base64 string (as requested)
CHAR_LIMIT = 32767

os.makedirs(output_folder, exist_ok=True)

def to_base64(byte_data: bytes) -> str:
    return base64.b64encode(byte_data).decode("utf-8")

def save_png_bytes(img: Image.Image) -> bytes:
    """Save image as PNG (optimized) to bytes."""
    buf = BytesIO()
    # Note: PNG ignores 'quality'; optimize helps a bit
    img.save(buf, format="PNG", optimize=True)
    return buf.getvalue()

def save_jpeg_bytes(img: Image.Image, quality: int) -> bytes:
    """Save image as JPEG to bytes; converts RGBA to RGB with white background."""
    if img.mode in ("RGBA", "LA"):
        bg = Image.new("RGB", img.size, (255, 255, 255))
        bg.paste(img, mask=img.split()[-1])
        work = bg
    else:
        work = img.convert("RGB")
    buf = BytesIO()
    work.save(buf, format="JPEG", quality=quality, optimize=True)
    return buf.getvalue()

def compress_to_limit(img: Image.Image, limit: int = CHAR_LIMIT) -> str:
    """
    Try PNG first, progressively downscaling until under limit.
    If still too large, fall back to JPEG with quality stepping
    (and additional downscaling if needed).
    Returns a base64 string.
    """
    # --- Strategy A: PNG + downscale ---
    w, h = img.size
    scale = 1.0
    work = img

    while True:
        png_bytes = save_png_bytes(work)
        b64 = to_base64(png_bytes)
        if len(b64) <= limit:
            return b64

        # Stop if already quite small; then try JPEG fallback
        if min(work.size) <= 64:
            break

        # Downscale by 10% and try again
        scale *= 0.9
        new_size = (max(1, int(w * scale)), max(1, int(h * scale)))
        work = img.resize(new_size, Image.LANCZOS)

    # --- Strategy B: JPEG + quality step + optional further downscale ---
    work = img
    scale = 1.0
    while True:
        for quality in range(85, 19, -5):  # 85, 80, ..., 20
            jpg_bytes = save_jpeg_bytes(work, quality)
            b64 = to_base64(jpg_bytes)
            if len(b64) <= limit:
                return b64

        # If not under limit, downscale further and try qualities again
        if min(work.size) <= 48:
            # We’ve tried hard enough
            raise ValueError("Cannot compress image under the character limit without excessive degradation.")

        scale *= 0.9
        new_size = (max(1, int(work.size[0] * scale)), max(1, int(work.size[1] * scale)))
        work = work.resize(new_size, Image.LANCZOS)

def is_png(path: str) -> bool:
    return os.path.splitext(path)[1].lower() == ".png"

# Collect all prior saved images (PNG files) from photos_folder
image_paths = [
    os.path.join(photos_folder, f)
    for f in os.listdir(photos_folder)
    if is_png(os.path.join(photos_folder, f))
]

output_path = os.path.join(output_folder, output_csv_file)

rows = []
for img_path in image_paths:
    try:
        with Image.open(img_path) as img:
            b64 = compress_to_limit(img, CHAR_LIMIT)
            file_name = os.path.basename(img_path)  # e.g., schew_panynj_gov.png
            rows.append((file_name, b64))
            print(f"✅ Encoded {file_name} (length={len(b64)})")
    except Exception as e:
        print(f"⚠️ Skipping {img_path}: {e}")

# Write CSV with required columns
with open(output_path, "w", newline="", encoding="utf-8") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["file_name", "base64"])  # header
    writer.writerows(rows)

print(f"Done. Wrote {len(rows)} rows to {output_path}")
```

```python
# photos_folder = dest_folder    # folder that contains your previously saved PNGs
# output_folder = base_folder      # folder where you want the CSV to go
# output_csv_file = "photos_base64.csv"   # CSV filename inside output_folder


def ensure_dirs():
    os.makedirs(photos_folder, exist_ok=True)
    os.makedirs(output_folder, exist_ok=True)

def filename_from_title(title: str) -> str:
    """
    Convert 'Last Name, First Name' -> 'lastname_firstname.png'
    - remove commas
    - replace spaces with underscores
    - lowercase for consistency
    """
    base = (title or "").strip()
    base = base.replace(",", "")
    base = "_".join(base.split())  # collapse whitespace -> underscores
    return base.lower() + ".png"

def to_base64(byte_data: bytes) -> str:
    return base64.b64encode(byte_data).decode("utf-8")

def save_png_bytes(img: Image.Image) -> bytes:
    buf = BytesIO()
    img.save(buf, format="PNG", optimize=True)
    return buf.getvalue()

def save_jpeg_bytes(img: Image.Image, quality: int) -> bytes:
    # Convert to RGB if needed (remove alpha)
    if img.mode in ("RGBA", "LA"):
        bg = Image.new("RGB", img.size, (255, 255, 255))
        bg.paste(img, mask=img.split()[-1])
        work = bg
    else:
        work = img.convert("RGB")
    buf = BytesIO()
    work.save(buf, format="JPEG", quality=quality, optimize=True)
    return buf.getvalue()

def compress_to_limit(img: Image.Image, limit: int = CHAR_LIMIT) -> str:
    """
    Try PNG first with progressive downscaling; if still too large,
    fall back to JPEG with quality steps and optional further downscale.
    Returns Base64 string.
    """
    # Strategy A: PNG + downscale
    w, h = img.size
    scale = 1.0
    work = img
    while True:
        png_bytes = save_png_bytes(work)
        b64 = to_base64(png_bytes)
        if len(b64) <= limit:
            return b64
        if min(work.size) <= 64:
            break
        scale *= 0.9
        new_size = (max(1, int(w * scale)), max(1, int(h * scale)))
        work = img.resize(new_size, Image.LANCZOS)

    # Strategy B: JPEG + quality stepping + possible further downscale
    work = img
    while True:
        for quality in range(85, 19, -5):  # 85, 80, ..., 20
            jpg_bytes = save_jpeg_bytes(work, quality)
            b64 = to_base64(jpg_bytes)
            if len(b64) <= limit:
                return b64

        if min(work.size) <= 48:
            raise ValueError("Cannot compress image under the character limit without excessive degradation.")

        new_w = max(1, int(work.size[0] * 0.9))
        new_h = max(1, int(work.size[1] * 0.9))
        work = work.resize((new_w, new_h), Image.LANCZOS)

# ========= MAIN =========
def main():
    ensure_dirs()
    
    # df = pd.read_csv(os.path.join(admin_data,"table_of_emails_contacts.csv"))["Employee.email"].tolist()
# df

# names = pd.read_csv(os.path.join(admin_data,"table_of_emails_contacts.csv"))["Employee.title"].tolist()
# names
    # 1) Load emails & names, keeping row alignment
    people = pd.read_csv(os.path.join(admin_data,"table_of_emails_contacts.csv"), dtype=str)[["Employee.email", "Employee.title"]].fillna("")

    # 2) Download photos using the email; save as lastname_firstname.png
    for _, row in people.iterrows():
        email = row["Employee.email"].strip()
        title = row["Employee.title"].strip()  # "Last, First"
        if not email or not title:
            print(f"⚠️ Missing email or name in row; skipping.")
            continue

        url = f"https://panynj.sharepoint.com/sites/WTCAnalyticsWarehouse/_layouts/15/userphoto.aspx?AccountName={email}"
        try:
            r = requests.get(url, cookies=cookies, timeout=20)
            if r.status_code == 200 and r.headers.get("Content-Type", "").startswith("image"):
                fn = filename_from_title(title)
                out_path = os.path.join(photos_folder, fn)
                with open(out_path, "wb") as f:
                    f.write(r.content)
                print(f"✅ Saved photo: {fn}")
            else:
                print(f"⚠️ No image for {email} ({r.status_code}); skipping save.")
        except Exception as e:
            print(f"⚠️ Error fetching {email}: {e}")

    # 3) Encode each saved photo to Base64 and collect rows: [name, base64]
    rows = []
    for _, row in people.iterrows():
        title = row["Employee.email"].strip()
        if not title:
            continue
        fn = filename_from_title(title)
        img_path = os.path.join(photos_folder, fn)
        if not os.path.exists(img_path):
            print(f"⚠️ Missing saved photo for '{title}' at {img_path}; skipping.")
            continue

        try:
            with Image.open(img_path) as img:
                b64 = compress_to_limit(img, CHAR_LIMIT)
                rows.append([title, b64])
                print(f"🔁 Encoded '{title}' (len={len(b64)})")
        except Exception as e:
            print(f"⚠️ Error encoding '{title}': {e}")

    # 4) Write the CSV with two columns: name, base64
    with open(output_csv_file, "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["name", "base64"])
        writer.writerows(rows)

    print(f"✅ Done. Wrote {len(rows)} rows to {output_csv_file}")

if __name__ == "__main__":
    main()
```
